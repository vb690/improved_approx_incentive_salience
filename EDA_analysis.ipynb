```python
import os 

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
GAMES = ['hmg', 'jc3', 'jc4', 'lis', 'lisbf', 'madness']
```

```python
for file in os.listdir('data\\train'):
    if file[-3:] == 'csv':
        df = pd.read_csv(f'data\\train\\{file}', 
                         converters={'activity_index_type': eval})
        print(df['activity_index_type'].apply(len).max())
```

```python
df = []
for game in GAMES:
    
    print(game)
    
    data = pd.read_csv(f'data\\csv\\cleaned\\{game}.csv')
    df.append(data)
    
df = pd.concat(df, ignore_index=True)
```

```python
len(df.user_id.unique())
```

```python
features = [
    'delta_sessions',
    'active_time',
    'session_time',
    'activity'
]
for game in GAMES:
    
    dat = df[df['context'] == game]
    
    for i in range(1, df['session_order'].max()):

        fig, axs = plt.subplots(1, 4, figsize=(15, 3))
        for ind, ax in enumerate(axs.flatten()):

            temporal_slice = dat[dat['session_order'] == i]
            temporal_slice = temporal_slice[features].values
            sns.distplot(temporal_slice[:, ind], ax=ax, kde=False)

        plt.tight_layout()
        plt.show()
```

```python
for session in df['session_order'].unique():
    
    fig, axs = plt.subplots(1, 1, figsize=(5, 5))
    sliced = df[df['session_order'] == session]
    corr = sliced[
        [
            'session_order', 
            'delta_sessions', 
            'active_time', 
            'session_time', 
            'activity'
        ]
    ].corr(method='spearman').values
    
    sns.heatmap(
        corr,
        vmin=-1,
        vmax=1,
        cmap='coolwarm',
        annot=True,
        ax=axs,
        cbar=False
    )

    plt.xticks([])
    plt.tight_layout()
    plt.show()
```
